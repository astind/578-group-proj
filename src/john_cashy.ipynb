```python
import pandas  
import matplotlib as plt
%matplotlib inline
filename="http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
 
df=pd.read_csv(filename,sep=",", header=None, names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'cash'])
```

```python
print(df)
```

```python
catcounts=df['cash'].value_counts()
ax = catcounts.plot(kind='pie')
# set the title
ax.set_title("Distribution of Wimbledon Victories 2015-2019")
```

```python
cros=pd.crosstab(df['cash'],df['race'])
```

```python
print(cros)
```

```python
cros.T.plot.pie(subplots=True, figsize=(22,12))
```

```python
cros.plot.pie(subplots=True, figsize=(22,12))
```

```python
cros=pd.crosstab(df['category'],df['gender'])
cros.T.plot.pie(subplots=True, figsize=(22,12))
```

```python
cros.plot.pie(subplots=True, figsize=(22,12))
```

```python
from statsmodels.graphics.mosaicplot import mosaic
mo=mosaic(df, ['race', 'cash'], title='DataFrame as Source')
plt.show()
 
```

```python
 
oldraces=df['race'].unique()
print(oldraces)
newraces=['White/Asian','Other','White/Asian','Other', 'Other']
df['racegroup'] = df['race'].replace(oldraces,newraces)
 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
fig, ax = plt.subplots(figsize=(18, 12))
ax=mosaic(df, ['racegroup', 'cash'],  gap=0.06, title='DataFrame',ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 12))
ax=mosaic(df, ['cash', 'racegroup'],  gap=0.06, title='DataFrame',ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18, 12))
ax=mosaic(df, ['cash', 'sex'],  gap=0.06, title='DataFrame',ax=ax)

plt.show()
```

```python
e=df['education'].unique()
ed_levels=['Y','N','N','Y','N','N','N','N','N','Y','Y','N','N','N','N','N']
df['college_degree'] = df['education'].replace(e,ed_levels)
cros=pd.crosstab(df['cash'],df['college_degree'])
cros.T.plot.pie(subplots=True, figsize=(22,12))
```
