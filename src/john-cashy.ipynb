```python
import pandas as pd  
import matplotlib as plt
%matplotlib inline
filename="http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
df=pd.read_csv(filename,sep=",", header=None, names=['age','employer','weight','degree','edu','marital','job','spouse','race','gender','gain','loss','hour','birth','category'])
```

```python
print(df)
```

```python
catcounts=df['category'].value_counts()
ax = catcounts.plot(kind='pie')
# set the title
ax.set_title("Distribution of Wimbledon Victories 2015-2019")
```

```python
cros=pd.crosstab(df['category'],df['race'])
```

```python
print(cros)
```

```python
cros.T.plot.pie(subplots=True, figsize=(22,12))
```

```python
cros.plot.pie(subplots=True, figsize=(22,12))
```

```python
cros=pd.crosstab(df['category'],df['gender'])
cros.T.plot.pie(subplots=True, figsize=(22,12))
```

```python
cros.plot.pie(subplots=True, figsize=(22,12))
```
